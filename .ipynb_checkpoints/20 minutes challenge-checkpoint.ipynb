```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import missingno as msno
```

```python
ratings = pd.read_csv("rating.csv")
```

```python
reviews = pd.read_csv('reviews.csv', index_col = 0)
```

```python
ratings.head(2)
```

```python
reviews.head(2)
```

```python
len(reviews.Name)-len(reviews.Name.drop_duplicates())
```

```python
len(ratings.Name)-len(ratings.Name.drop_duplicates())
```

```python
merged_df = pd.merge(reviews,ratings, on=['Datetime','Name','OverallRating']).drop(columns='index')
```

```python
merged_df.head(2)
```

```python
len(merged_df.Name)-len(merged_df.Name.drop_duplicates())
```

```python
merged_df.describe().iloc[1].sort_values(ascending=False).plot(kind='barh')
```

```python
clean_df = merged_df.drop(columns=['Aircraft', 'Wifi&Connectivity', 'InflightEntertainment'])
```

```python
clean_df.isna().sum() / clean_df.shape[1]
```

```python
msno.matrix(clean_df)
```

```python
clean_df.dropna(subset = ['OverallRating', 'TypeOfTraveller', 'SeatType', 'DateFlown', 'Route'], inplace=True)
```

```python
clean_df.info()
```

```python
clean_df.head(2)
```

```python
clean_df.TypeOfTraveller.value_counts().is_unique
```

```python
clean_df.dtypes
```

```python
clean_df['Datetime'] = pd.to_datetime(clean_df['Datetime'])
```

```python
clean_df['month'] = clean_df['Datetime'].dt.month
```

```python
clean_df.head(2)
```

```python
clean_df.groupby(['month']).OverallRating.mean().plot(kind='barh')
```

```python
clean_df.groupby(['SeatType']).OverallRating.mean().plot(kind='barh')
```

```python
clean_df.columns
```

```python
clean_df.dtypes
```

```python
clean_df["VerifiedReview"] = clean_df["VerifiedReview"].astype(int)
```

```python
clean_df_corr_matrix = clean_df.drop(columns=['Datetime','Name', 'ReviewHeader', 'ReviewBody', 'TypeOfTraveller', 'SeatType', 'Route', 'DateFlown', 'Recommended']).corr()
```

```python
sns.heatmap(clean_df_corr_matrix, annot=True)
```
